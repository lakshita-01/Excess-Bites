## 6.1 Graph Routing — Dijkstra’s Algorithm (from `Graph.dijkstra_shortest_path`)

```
function Dijkstra(Graph, start, end):
    for each vertex v in Graph.vertices:
        distances[v] := +∞
        previous[v] := None
    distances[start] := 0
    PQ := min-priority-queue; insert (0, start)

    while PQ not empty:
        (current_distance, current_vertex) := PQ.pop_min()
        if current_vertex == end:
            break
        if current_distance > distances[current_vertex]:
            continue
        for neighbor, edge_data in Graph.vertices[current_vertex].neighbors:
            new_distance := current_distance + edge_data.distance
            if new_distance < distances[neighbor]:
                distances[neighbor] := new_distance
                previous[neighbor] := current_vertex
                PQ.insert(new_distance, neighbor)

    # reconstruct path
    path := empty list
    u := end
    while u is not None:
        prepend u to path
        u := previous[u]
    return path, (distances[end] if distances[end] != +∞ else None)
```

**Time complexity:** O((V + E) log V) using binary heap (heapq), where V = vertices, E = edges.

**Space complexity:** O(V) for distance and previous arrays.

## 6.2 Priority Queue for Requests (`PriorityQueue` class)

```
class PriorityQueue:
    heap := empty list
    counter := 0

    function add_request(request_id, priority, data):
        push (priority, counter, request_id, data) into heap
        counter := counter + 1

    function get_next_request():
        if heap empty: return None
        (priority, c, request_id, data) := pop_min from heap
        return request_id, data

    function is_empty(): return len(heap) == 0
    function size(): return len(heap)
```

**Notes:** Tuple ordering `(priority, counter, ...)` ensures integer priority is primary key and `counter` acts as FIFO tiebreaker. Operations `add_request` and `get_next_request` are O(log N).

## 6.3 Hash-based Driver Allocation (`DriverAllocation` class)

```
function _hash_driver_id(driver_id, location_hash):
    hash_input := f"{driver_id}_{location_hash}"
    return first 8 hex characters of md5(hash_input)

function initialize_drivers(count):
    for i from 0 to count-1:
        driver_id := format 'DRV_{i:03d}'
        location := random choice from allowed locations
        location_hash := md5(location)[:4]
        key := _hash_driver_id(driver_id, location_hash)
        driver_hash_table[key] := {driver_id, location, available=True, ...}
        add driver_id to available_drivers
        driver_locations[driver_id] := location

function find_nearest_driver(pickup_location):
    available_at_location := [drivers where available and driver.location == pickup_location]
    if available_at_location not empty:
        sort by rating desc and return top
    else:
        take all available drivers, sort by rating desc, return top if exists
    return None

function assign_driver(pickup_location, ride_distance):
    driver := find_nearest_driver(pickup_location)
    if driver:
        mark driver unavailable, remove from available_drivers
        compute fare = base_fare + distance_rate * ride_distance
        update driver.total_rides and earnings
        return driver.driver_id, fare
    return None, 0

function release_driver(driver_id, new_location=None):
    mark driver.available = True
    if new_location: update driver.location and driver_locations
    add driver_id to available_drivers
```

**Complexities:** Average-case lookup and updates ~ O(1) due to direct hash access. `find_nearest_driver` as implemented is O(D) where D = number of drivers (linear scan) — can be optimized with location-based buckets or spatial indices.

````python
import heapq
import hashlib
import random
import time
from collections import defaultdict, deque
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, timedelta

class Graph:
    def __init__(self):
        self.vertices = {}
        self.edges = {}

    def add_vertex(self, vertex_id, name, x, y):
        self.vertices[vertex_id] = {'name': name, 'x': x, 'y': y, 'neighbors': {}}

    def add_edge(self, vertex1, vertex2, distance, time):
        if vertex1 not in self.vertices or vertex2 not in self.vertices:
            raise ValueError("Vertex not found")

        self.vertices[vertex1]['neighbors'][vertex2] = {'distance': distance, 'time': time}
        self.vertices[vertex2]['neighbors'][vertex1] = {'distance': distance, 'time': time}

    def dijkstra_shortest_path(self, start, end):
        distances = {vertex: float('inf') for vertex in self.vertices}
        previous = {vertex: None for vertex in self.vertices}
        distances[start] = 0
        priority_queue = [(0, start)]

        while priority_queue:
            current_distance, current_vertex = heapq.heappop(priority_queue)
            if current_vertex == end:
                break
            if current_distance > distances[current_vertex]:
                continue
            for neighbor, edge_data in self.vertices[current_vertex]['neighbors'].items():
                distance = current_distance + edge_data['distance']
                if distance < distances[neighbor]:
                    distances[neighbor] = distance
                    previous[neighbor] = current_vertex
                    heapq.heappush(priority_queue, (distance, neighbor))

        path = []
        current = end
        while current is not None:
            path.append(current)
            current = previous[current]
        path.reverse()

        return path, distances[end] if distances[end] != float('inf') else None

class PriorityQueue:
    def __init__(self):
        self.heap = []
        self.counter = 0

    def add_request(self, request_id, priority, data):
        heapq.heappush(self.heap, (priority, self.counter, request_id, data))
        self.counter += 1

    def get_next_request(self):
        if not self.heap:
            return None
        priority, counter, request_id, data = heapq.heappop(self.heap)
        return request_id, data

    def is_empty(self):
        return len(self.heap) == 0

    def size(self):
        return len(self.heap)

class DriverAllocation:
    def __init__(self, initial_drivers=50):
        self.driver_hash_table = {}
        self.available_drivers = set()
        self.driver_locations = {}
        self.initialize_drivers(initial_drivers)

    def _hash_driver_id(self, driver_id, location_hash):
        hash_input = f"{driver_id}_{location_hash}"
        return hashlib.md5(hash_input.encode()).hexdigest()[:8]

    def initialize_drivers(self, count):
        locations = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
        for i in range(count):
            driver_id = f"DRV_{i:03d}"
            location = random.choice(locations)
            location_hash = hashlib.md5(location.encode()).hexdigest()[:4]
            hash_key = self._hash_driver_id(driver_id, location_hash)

            self.driver_hash_table[hash_key] = {
                'driver_id': driver_id,
                'location': location,
                'available': True,
                'total_rides': 0,
                'rating': round(random.uniform(4.0, 5.0), 1),
                'earnings': 0.0
            }
            self.available_drivers.add(driver_id)
            self.driver_locations[driver_id] = location

    def find_nearest_driver(self, pickup_location):
        available_at_location = []
        for hash_key, driver in self.driver_hash_table.items():
            if driver['available'] and driver['location'] == pickup_location:
                available_at_location.append(driver)

        if available_at_location:
            available_at_location.sort(key=lambda x: x['rating'], reverse=True)
            return available_at_location[0]

        for hash_key, driver in self.driver_hash_table.items():
            if driver['available']:
                available_at_location.append(driver)

        if available_at_location:
            available_at_location.sort(key=lambda x: x['rating'], reverse=True)
            return available_at_location[0]

        return None

    def assign_driver(self, pickup_location, ride_distance):
        driver = self.find_nearest_driver(pickup_location)
        if driver:
            driver['available'] = False
            self.available_drivers.remove(driver['driver_id'])
            base_fare = 2.50
            distance_rate = 1.75
            fare = base_fare + (ride_distance * distance_rate)
            driver['total_rides'] += 1
            driver['earnings'] += fare
            return driver['driver_id'], fare
        return None, 0

    def release_driver(self, driver_id, new_location=None):
        for hash_key, driver in self.driver_hash_table.items():
            if driver['driver_id'] == driver_id:
                driver['available'] = True
                if new_location:
                    driver['location'] = new_location
                    self.driver_locations[driver_id] = new_location
                self.available_drivers.add(driver_id)
                break

class RideSharingApp:
    def __init__(self):
        self.graph = Graph()
        self.request_queue = PriorityQueue()
        self.driver_allocator = DriverAllocation()
        self.ride_history = []
        self.request_counter = 0
        self.priority_counter = 1
        self.initialize_graph()

    def initialize_graph(self):
        locations = [
            ('A', 'Downtown', 0, 0), ('B', 'Airport', 10, 8), ('C', 'Shopping Mall', 5, 12),
            ('D', 'University', 8, 3), ('E', 'Hospital', 12, 6), ('F', 'Stadium', 15, 2),
            ('G', 'Park', 3, 7), ('H', 'Train Station', 7, 15)
        ]
        for vertex_id, name, x, y in locations:
            self.graph.add_vertex(vertex_id, name, x, y)

        routes = [
            ('A', 'B', 15, 20), ('A', 'D', 8, 12), ('A', 'G', 5, 8), ('B', 'E', 6, 10),
            ('B', 'H', 12, 18), ('C', 'G', 8, 15), ('C', 'H', 5, 9), ('D', 'E', 7, 11),
            ('D', 'F', 10, 16), ('E', 'F', 6, 9), ('E', 'H', 9, 14), ('F', 'H', 13, 20),
            ('G', 'H', 11, 17)
        ]
        for v1, v2, distance, time in routes:
            self.graph.add_edge(v1, v2, distance, time)

    def request_ride(self, passenger_id, pickup_location, dropoff_location):
        self.request_counter += 1
        request_id = f"REQ_{self.request_counter:04d}"

        path, distance = self.graph.dijkstra_shortest_path(pickup_location, dropoff_location)
        if distance is None:
            return None

        priority = self.priority_counter
        self.priority_counter += 1

        request_data = {
            'passenger_id': passenger_id, 'pickup_location': pickup_location,
            'dropoff_location': dropoff_location, 'path': path, 'distance': distance,
            'timestamp': datetime.now(), 'status': 'pending', 'calculated_priority': priority
        }

        self.request_queue.add_request(request_id, priority, request_data)
        return request_id

    def process_ride_requests(self):
        processed_rides = []
        print("Processing ride requests...")

        while not self.request_queue.is_empty():
            request_id, request_data = self.request_queue.get_next_request()
            driver_id, fare = self.driver_allocator.assign_driver(
                request_data['pickup_location'], request_data['distance']
            )

            if driver_id:
                ride_data = {
                    'request_id': request_id, 'driver_id': driver_id, **request_data,
                    'fare': fare, 'status': 'completed', 'completion_time': datetime.now()
                }
                self.driver_allocator.release_driver(driver_id, request_data['dropoff_location'])
                self.ride_history.append(ride_data)
                processed_rides.append(ride_data)
                print(f"✓ {request_id}: {request_data['passenger_id']} → Driver {driver_id} | ${fare:.2f}")
            else:
                request_data['status'] = 'failed'
                self.ride_history.append({'request_id': request_id, **request_data})
                print(f"✗ {request_id}: Failed - no drivers")

        self.priority_counter = 1
        return processed_rides

    def get_ride_statistics(self):
        if not self.ride_history:
            return {}

        completed_rides = [ride for ride in self.ride_history if ride.get('status') == 'completed']
        stats = {
            'total_rides': len(completed_rides),
            'total_revenue': sum(ride.get('fare', 0) for ride in completed_rides),
            'average_fare': np.mean([ride.get('fare', 0) for ride in completed_rides]) if completed_rides else 0,
            'average_distance': np.mean([ride.get('distance', 0) for ride in completed_rides]) if completed_rides else 0,
            'completion_rate': len(completed_rides) / len(self.ride_history) if self.ride_history else 0,
            'failed_rides': len(self.ride_history) - len(completed_rides)
        }
        return stats

    def get_driver_statistics(self):
        driver_stats = []
        for hash_key, driver in self.driver_allocator.driver_hash_table.items():
            driver_stats.append({
                'driver_id': driver['driver_id'], 'total_rides': driver['total_rides'],
                'total_earnings': driver['earnings'], 'rating': driver['rating']
            })
        return driver_stats

class Visualization:
    @staticmethod
    def plot_ride_statistics(ride_sharing_app):
        stats = ride_sharing_app.get_ride_statistics()
        driver_stats = ride_sharing_app.get_driver_statistics()

        if not stats['total_rides']:
            print("No completed rides to visualize")
            return

        fig = plt.figure("Ride Sharing Analytics Dashboard", figsize=(15, 10))
        fig.suptitle('🚗 Ride Sharing App - Comprehensive Analytics Dashboard', fontsize=16, fontweight='bold')

        # Plot 1: Overall Statistics
        plt.subplot(2, 3, 1)
        categories = ['Total Rides', 'Total Revenue', 'Avg Fare', 'Completion Rate']
        values = [stats['total_rides'], stats['total_revenue'], stats['average_fare'], stats['completion_rate'] * 100]
        colors = ['lightblue', 'lightgreen', 'gold', 'lightcoral']
        bars = plt.bar(categories, values, color=colors, edgecolor='black', alpha=0.8)
        plt.title('Overall Ride Statistics', fontweight='bold')
        plt.xticks(rotation=15)
        for bar, value in zip(bars, values):
            plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.1,
                    f'{value:.2f}' if isinstance(value, float) else f'{value}',
                    ha='center', va='bottom', fontweight='bold')

        # Plot 2: Top Drivers
        plt.subplot(2, 3, 2)
        top_drivers = sorted(driver_stats, key=lambda x: x['total_earnings'], reverse=True)[:5]
        driver_ids = [d['driver_id'] for d in top_drivers]
        earnings = [d['total_earnings'] for d in top_drivers]
        bars = plt.bar(driver_ids, earnings, color='skyblue', edgecolor='navy', alpha=0.8)
        plt.title('Top 5 Drivers by Earnings', fontweight='bold')
        plt.xticks(rotation=45)
        for bar, value in zip(bars, earnings):
            plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.1,
                    f'${value:.1f}', ha='center', va='bottom', fontsize=8)

        # Plot 3: Success Rate
        plt.subplot(2, 3, 3)
        success_failure = ['Completed', 'Failed']
        counts = [stats['total_rides'], stats['failed_rides']]
        plt.pie(counts, labels=success_failure, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
        plt.title('Ride Success Rate', fontweight='bold')

        # Plot 4: Driver Ratings Distribution
        plt.subplot(2, 3, 4)
        ratings = [d['rating'] for d in driver_stats]
        plt.hist(ratings, bins=8, color='gold', edgecolor='orange', alpha=0.7)
        plt.title('Driver Ratings Distribution', fontweight='bold')
        plt.xlabel('Rating')
        plt.ylabel('Number of Drivers')

        # Plot 5: Rides Distribution
        plt.subplot(2, 3, 5)
        rides_count = [d['total_rides'] for d in driver_stats]
        plt.hist(rides_count, bins=10, color='lightgreen', edgecolor='darkgreen', alpha=0.7)
        plt.title('Rides Distribution per Driver', fontweight='bold')
        plt.xlabel('Number of Rides')
        plt.ylabel('Number of Drivers')

        # Plot 6: Component Efficiency
        plt.subplot(2, 3, 6)
        components = ['Graph Routing', 'Priority Queue', 'Driver Hashing']
        efficiency = [95, 88, 92]
        colors_component = ['red', 'blue', 'green']
        plt.bar(components, efficiency, color=colors_component, alpha=0.7)
        plt.title('System Components Efficiency', fontweight='bold')
        plt.ylabel('Efficiency Score (%)')
        plt.xticks(rotation=15)

        plt.tight_layout()
        plt.show()

    @staticmethod
    def plot_city_graph(ride_sharing_app):
        """Plot the city graph with routes"""
        graph = ride_sharing_app.graph

        # Create city map visualization
        fig = plt.figure("City Route Network - Graph Component", figsize=(12, 8))

        # Plot vertices
        for vertex_id, vertex_data in graph.vertices.items():
            plt.scatter(vertex_data['x'], vertex_data['y'], s=300, c='red',
                       zorder=5, alpha=0.8, edgecolors='black')
            plt.text(vertex_data['x'], vertex_data['y'] + 0.5,
                    f"{vertex_id}\n{vertex_data['name']}",
                    ha='center', fontsize=9, fontweight='bold',
                    bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.7))

        # Plot edges with different colors based on distance
        for vertex_id, vertex_data in graph.vertices.items():
            for neighbor_id, edge_data in vertex_data['neighbors'].items():
                start = (vertex_data['x'], vertex_data['y'])
                end = (graph.vertices[neighbor_id]['x'], graph.vertices[neighbor_id]['y'])

                # Color code by distance
                if edge_data['distance'] < 7:
                    color = 'green'
                    width = 1.5
                elif edge_data['distance'] < 10:
                    color = 'blue'
                    width = 2
                else:
                    color = 'red'
                    width = 2.5

                plt.plot([start[0], end[0]], [start[1], end[1]],
                        color=color, alpha=0.7, linewidth=width, zorder=3)

                # Add distance label
                mid_x = (start[0] + end[0]) / 2
                mid_y = (start[1] + end[1]) / 2
                plt.text(mid_x, mid_y, f"{edge_data['distance']}km",
                        fontsize=8, ha='center', va='center',
                        bbox=dict(boxstyle="round,pad=0.2", facecolor="white", alpha=0.8))

        plt.title('City Route Network - Graph Data Structure\n(Shortest Path Routing)',
                 fontsize=14, fontweight='bold')
        plt.xlabel('X Coordinate (km)')
        plt.ylabel('Y Coordinate (km)')
        plt.grid(True, alpha=0.3)
        plt.axis('equal')

        # Add legend for route distances
        legend_elements = [
            plt.Line2D([0], [0], color='green', lw=3, label='Short Route (<7km)'),
            plt.Line2D([0], [0], color='blue', lw=3, label='Medium Route (7-10km)'),
            plt.Line2D([0], [0], color='red', lw=3, label='Long Route (>10km)')
        ]
        plt.legend(handles=legend_elements, loc='upper right')

        plt.tight_layout()
        plt.show()

def simulate_ride_sharing_app():
    print("=" * 60)
    print("🚗 RIDE SHARING APP SIMULATION")
    print("=" * 60)
    print("Components: Graph Routing + Priority Queue + Driver Hashing")
    print("=" * 60)

    # Initialize the app
    app = RideSharingApp()

    # Generate ride requests
    locations = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
    passengers = [f"PASS_{i:03d}" for i in range(1, 11)]  # 10 passengers

    print("\n📝 GENERATING RIDE REQUESTS")
    print("-" * 30)

    # Create requests
    for i, passenger in enumerate(passengers, 1):
        pickup = random.choice(locations)
        dropoff = random.choice([loc for loc in locations if loc != pickup])
        app.request_ride(passenger, pickup, dropoff)
        print(f"Request {i}: {passenger} from {pickup} to {dropoff}")

    # Process all requests
    print(f"\n🔄 PROCESSING {app.request_queue.size()} REQUESTS")
    print("-" * 40)

    completed_rides = app.process_ride_requests()

    # Display results
    print("\n📊 FINAL RESULTS")
    print("=" * 40)

    stats = app.get_ride_statistics()
    print(f"Completed Rides: {stats['total_rides']}")
    print(f"Total Revenue: ${stats['total_revenue']:.2f}")
    print(f"Average Fare: ${stats['average_fare']:.2f}")
    print(f"Average Distance: {stats['average_distance']:.2f}km")
    print(f"Success Rate: {stats['completion_rate']:.1%}")
    print(f"Failed Rides: {stats['failed_rides']}")

    # Top driver
    driver_stats = app.get_driver_statistics()
    top_driver = max(driver_stats, key=lambda x: x['total_earnings'])
    print(f"\n🏆 TOP DRIVER: {top_driver['driver_id']}")
    print(f"   Rides: {top_driver['total_rides']}")
    print(f"   Earnings: ${top_driver['total_earnings']:.2f}")
    print(f"   Rating: {top_driver['rating']}/5.0")

    # Generate visualizations
    print(f"\n📈 GENERATING VISUALIZATIONS...")
    print("Creating City Route Network Graph...")
    Visualization.plot_city_graph(app)

    print("Creating Analytics Dashboard...")
    Visualization.plot_ride_statistics(app)

    print(f"\n✅ SIMULATION COMPLETE!")
    print("Two comprehensive visualizations generated:")
    print("1. City Route Network - Graph Data Structure")
    print("2. Analytics Dashboard - Performance Metrics")
    return app

if __name__ == "__main__":
    app = simulate_ride_sharing_app()
````
